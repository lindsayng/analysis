# Defining the Dataframe

```python
import csv
import pandas as pd
from datetime import datetime
import postpatch_funcs as pf
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='white', color_codes = True)
%matplotlib inline
```

```python
initial = pd.read_csv('C:/users/lindsayn/documents/stash/ps-metadata-analysis/data/patch_seq_log_mouse.csv')
```

```python
ps = initial.loc[:,["User","Date", "Post patch?", "Post patch pipette R", "Internal version", "PCR cycles", "SM_QC_PF", "Resolution index", "Res index pctl.", "Bad dates"]]
ps.dropna(subset = ["Post patch?"], inplace = True)
ps["Date"] = pd.to_datetime(ps["Date"], format = '%y%m%d', errors = 'coerce')
ps = ps[ps["PCR cycles"] == 21]
ps = ps[ps["SM_QC_PF"] != "na"]
ps = ps[ps["SM_QC_PF"] != "smarter_fail"]
ps.dropna(subset = ["SM_QC_PF"], inplace=True)
ps = pf.postpatch_reclass(ps)
ps.sort_values("Date")
```

```python
ps = ps.loc[:,["Internal version","SM_QC_PF", "Date"]]
ps = ps.groupby(["Internal version","SM_QC_PF"]).count()
ps = ps.reset_index()
ps = ps.rename(columns = {"SM_QC_PF": ''})
ps = ps.pivot_table(values=[''], index=['Internal version'], columns=[''], aggfunc='sum')
ps.fillna(0,inplace =True)
```

```python
ps = ps["Date"]
ps = ps.loc[:,["pass", "fail"]]
ps2 = ps.div(ps.sum(1), axis=0)*100
ps
```

```python
colors = sns.color_palette("Blues_r", 2)
ax1 = ps2.plot(kind= "bar", stacked = True, figsize=(10,7),fontsize = 14, color = colors)
plt.title("Does alexa-488 internal affect SM QC pass percentage?", fontsize= 22, loc = "center")
plt.legend(["SM QC Pass", "SM QC Fail"] ,loc = 'center right',labelspacing=-2.3, fontsize =14)
plt.ylabel("Percentage of Cells", fontsize= 14)
plt.xlabel("Internal Version", fontsize = 14)
plt.xlim(-0.5,2)
plt.ylim(0,119)
plt.xticks(rotation=0)
sns.despine()

rects = ax1.patches
totals = ps.sum(1)
labels = totals
labels = pd.to_numeric(labels, errors='raise', downcast='integer')
for rect, label in zip (rects, labels):
    y_value = 101
    x_value = rect.get_x() + rect.get_width() / 2
    ax1.text(x_value, y_value, label, ha='center', va='bottom', size=13)
    
rects = ax1.patches
totals = ps["pass"]
labels = totals
labels = pd.to_numeric(labels, errors='raise', downcast='integer')
for rect, label in zip (rects, labels):
    y_value = 40
    x_value = rect.get_x() + rect.get_width() / 2
    ax1.text(x_value, y_value, label, ha='center', va='bottom', size=13)
    
x1, x2 = 0, 1   
y = 1
plt.text((x1+x2)*.5, y+110, "ns, p = 0.85", ha='center', va='bottom', size=13)

plt.savefig("C:/users/lindsayn/documents/github/analysis/figures/SMQC_standard_vs_alexa", dpi=100)
```
